```python
import yfinance as yf
import pandas as pd
import tqdm
```

```python
tickers = pd.read_csv("./2023-01-05-tickers.csv").Symbols

tickers
```

```python
market_caps = {}

for i, t in tqdm.tqdm(enumerate(tickers)):
    try:
        mc = yf.Ticker(t).info["marketCap"]
    except Exception as e:
        try:
            mc = yf.Ticker(t).info["totalAssets"]
        except Exception as e2:
            continue
    #print(mc)
    market_caps[i] = mc
```
